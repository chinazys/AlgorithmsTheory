```python
import numpy as np
import matplotlib.pyplot as plt
from searchers.binary_search import binary_search
from searchers.interpolation_search import interpolation_search

def f1(x):
    return 14 * x

def f2(x):
    return 100 * np.log(x)

def f3(x):
    return 0.03 * x ** 3
```

```python
interval = [1, 20]
num_points = 100

x_values = np.linspace(interval[0], interval[1], num_points)

y_values1 = f1(x_values)
y_values2 = f2(x_values)
y_values3 = f3(x_values)

target_values = np.linspace(interval[0], interval[1], num_points)

binary_results1 = []
binary_results2 = []
binary_results3 = []

for target in target_values:
    binary_results1.append(binary_search(y_values1, target))
    binary_results2.append(binary_search(y_values2, target))
    binary_results3.append(binary_search(y_values3, target))

interpolation_results1 = []
interpolation_results2 = []
interpolation_results3 = []

for target in target_values:
    interpolation_results1.append(interpolation_search(y_values1, target))
    interpolation_results2.append(interpolation_search(y_values2, target))
    interpolation_results3.append(interpolation_search(y_values3, target))

plt.figure(figsize=(16, 20))
plt.subplot(3, 1, 1)

plt.ylim([0, 20])
plt.plot(target_values, binary_results1, 'red', label='Binary Search')
plt.plot(target_values, interpolation_results1, 'green', label='Interpolation Search')
plt.xlabel('Search Target')
plt.ylabel('Number of Operations')
plt.title('f(x) = 14 * x')
plt.legend()

plt.subplot(3, 1, 2)

plt.ylim([0, 20])
plt.plot(target_values, binary_results2, 'red', label='Binary Search')
plt.plot(target_values, interpolation_results2, 'green', label='Interpolation Search')
plt.xlabel('Search Target')
plt.ylabel('Number of Operations')
plt.title('\nf(x) = 100 * ln(x)')
plt.legend()

plt.subplot(3, 1, 3)

plt.ylim([0, 20])
plt.plot(target_values, binary_results3, 'red', label='Binary Search')
plt.plot(target_values, interpolation_results3, 'green', label='Interpolation Search')
plt.xlabel('Search Target')
plt.ylabel('Number of Operations')
plt.title('\nf(x) = 0.03 * x^3')
plt.legend()

plt.tight_layout()
plt.show()
```
