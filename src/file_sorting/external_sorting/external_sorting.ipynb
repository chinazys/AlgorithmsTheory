```python
import numpy as np
import matplotlib.pyplot as plt
from sorters.balanced_multiway_sorter import BalancedMultiwaySorter

ARRAY_SIZE = 3000
BUFFER_SIZE = 50
FILES_QUANTITY = 8
INPUT_FILENAME = 'unsorted.csv'
OUTPUT_FILENAME = 'sorted.csv'
```

```python
unsorted_array = np.arange(ARRAY_SIZE)
np.random.shuffle(unsorted_array)
np.savetxt(INPUT_FILENAME, unsorted_array, fmt = '%1.0f')
```

```python
files_quantity_test = [2 * v for v in range(2, 16)]
files_quantity_time = []

for files_quantity in files_quantity_test:
    balanced_multiway_sorter = BalancedMultiwaySorter(INPUT_FILENAME, BUFFER_SIZE, files_quantity)
    balanced_multiway_sorter.run(OUTPUT_FILENAME)
    files_quantity_time.append(balanced_multiway_sorter.execution_time)

buffer_size_test = [10 * v for v in range(1, 6)]
buffer_size_time = []

for buffer_size in buffer_size_test:
    balanced_multiway_sorter = BalancedMultiwaySorter(INPUT_FILENAME, buffer_size, FILES_QUANTITY)
    balanced_multiway_sorter.run(OUTPUT_FILENAME)
    buffer_size_time.append(balanced_multiway_sorter.execution_time)
print(buffer_size_time)

plt.rcParams["figure.figsize"] = (16, 8)
figure, axis = plt.subplots(1, 2)

axis[0].plot(files_quantity_test, files_quantity_time, color='r')
axis[0].set_xlabel('Files Quantity')
axis[0].set_ylabel('Execution Time')
axis[0].set_ylim((0, 1.2 * max(files_quantity_time)))

axis[1].plot(buffer_size_test, buffer_size_time, color='r')
axis[1].set_xlabel('Buffer Size')
axis[1].set_ylabel('Execution Time')
axis[1].set_ylim((0, 1.2 * max(buffer_size_time)))

plt.show()
```
