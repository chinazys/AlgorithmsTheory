```python
import matplotlib.pyplot as plt

from util.array_type_enums import *
from util.array_union import ArrayUnion
from test_sorters.verify_sorters import verify_all_sorters
from test_sorters.measure_sorters import measure_all_sorters

MIN_ARRAY_LENGTH = 2
MAX_ARRAY_LENGTH = 25000
```

```python
verify_all_sorters()
```

```python
def compare_sorters_on_same_array(array_type, axis, plot_x, plot_y, plot_title):
    n_measures, t_measures = measure_all_sorters(array_type, MIN_ARRAY_LENGTH, MAX_ARRAY_LENGTH)

    axis[plot_x][plot_y].plot(n_measures, t_measures[0], color='green', label='LSD Sorter (base = 2)')
    axis[plot_x][plot_y].plot(n_measures, t_measures[1], color='cyan', label='LSD Sorter (base = 10)')
    axis[plot_x][plot_y].plot(n_measures, t_measures[2], color='blue', label='LSD Sorter (base = 16)')
    axis[plot_x][plot_y].plot(n_measures, t_measures[3], color='yellow', label='MSD Sorter (base = 2)')
    axis[plot_x][plot_y].plot(n_measures, t_measures[4], color='orange', label='MSD Sorter (base = 10)')
    axis[plot_x][plot_y].plot(n_measures, t_measures[5], color='red', label='MSD Sorter (base = 16)')
    axis[plot_x][plot_y].plot(n_measures, t_measures[6], color='black', label='Heap Sorter')
    axis[plot_x][plot_y].plot(n_measures, t_measures[7], color='brown', label='Bentley Sorter')
    axis[plot_x][plot_y].plot(n_measures, t_measures[8], color='grey', label='Default Sorter')

    axis[plot_x][plot_y].set_title(plot_title)
    axis[plot_x][plot_y].legend()
```

```python
plt.rcParams["figure.figsize"] = (16, 10)
figure, axis = plt.subplots(2, 2)

compare_sorters_on_same_array(ARRAY_RANDOM, axis, 0, 0, "RANDOM ARRAY")
compare_sorters_on_same_array(ARRAY_ASCENDING, axis, 0, 1, "ASCENDING ARRAY")
compare_sorters_on_same_array(ARRAY_DESCENDING, axis, 1, 0, "DESCENDING ARRAY")
compare_sorters_on_same_array(ARRAY_IDENTICAL, axis, 1, 1, "IDENTICAL ARRAY")
```
